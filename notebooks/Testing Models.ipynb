Loading data set dor initial tests

```python
#loading all data 

#loading trains 
data_train = pd.read_csv('../raw_data/sales_train.csv')

#loading test data 
data_test = pd.read_csv('../raw_data/test.csv')

#loading item categories data 
item_categories = pd.read_csv('../raw_data/item_categories.csv')

#loading items
items = pd.read_csv('../raw_data/items.csv')

#loading shops 
shops = pd.read_csv('../raw_data/shops.csv')

#loadinf sub
sample = pd.read_csv('../raw_data/sample_submission.csv')


```

```python
data_train.sort_values('date', ascending=True)
```

```python
pivot = pd.pivot_table(data_train, values='item_cnt_day', index=['shop_id','item_id'],
                    columns=['date_block_num'], aggfunc=np.sum)
```

```python
pivot.head()
```

```python
pivot = pivot.fillna(0)
```

```python
baseline = pivot
```

```python
baseline['average'] = baseline.mean(axis=1)
```

```python
baseline
```

Checking missing values

```python
baseline = baseline.reset_index()
```

```python
baseline
```

```python
baseline['id_final'] = baseline.apply(lambda x: f"{x['shop_id']}-{x['item_id']}", axis=1)
```

```python
baseline
```

```python
data_test.info()
```

```python
data_test['id_final'] = data_test.apply(lambda x: f"{x['shop_id']}-{x['item_id']}", axis=1)
```

```python
baseline_test = pd.merge(data_test, baseline, on="id_final", how="left")
```

```python
final = baseline_test[['ID','average']]
```

```python
final = final.fillna(0)
```

```python
final = final.rename(columns={'average': 'item_cnt_month'})
```

```python
final.sort_values('ID', ascending=True)
```

```python
with open('baseline2.csv', 'w') as csv_file:
    final.to_csv(path_or_buf=csv_file, index=False)
```
